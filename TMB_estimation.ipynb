```python
import pandas as pd
import numpy as np
from sklearn.metrics import r2_score
import joblib
```

```python
mutmtx = pd.read_csv('hub_mutdata.csv', index_col=0)
```

```python
mutmtx
```

```python
mut_load = pd.read_csv('mut_load.csv', index_col=0)
mut_load
```

```python
model = joblib.load('Lasso Regression.pkl')
```

```python
imp_coef = pd.Series(model.coef_, index = model.feature_names_in_)
```

```python
imp_coef
```

```python
x = mutmtx[mutmtx.index.isin(imp_coef.index)]
x = x.transpose()
y_pred = model.predict(x)
mut_load['Lasso'] = y_pred
```
