```python
import pandas as pd
import numpy as np
import joblib
```

```python
mutmtx = pd.read_csv('pancancer_mutdata.csv', index_col=0)
```

```python
mutmtx
```

```python
mut_load = pd.read_csv('pancancer_mutload.csv', index_col=0)
mut_load
```

```python
model = joblib.load('Lasso Regression.pkl')
```

```python
imp_coef = pd.Series(model.coef_, index = model.feature_names_in_)
```

```python
imp_coef
```

```python
x = mutmtx[mutmtx.index.isin(imp_coef.index)]
x = x.transpose()
y_pred = model.predict(x)
```

```python
mut_load['20 gene TMB'] = y_pred
```

```python
mut_load
```

```python
mut_load.to_csv('Pancancer_20_gene_TMB.csv')
```
